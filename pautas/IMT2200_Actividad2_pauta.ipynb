```python
import pandas as pd
import requests

url='https://github.com/paguirre-uc/imt2200_data/blob/main/data_1661435089.xlsx?raw=true'

r=requests.get(url)
```

```python
open('data_a2.xlsx','wb').write(r.content)
```

```python
#leemos el archivo local usando pandas
df=pd.read_excel('data_a2.xlsx',engine='openpyxl',skiprows=0)
df

#### Hasta acá: nota 5.0
```

```python
paises=df['País__ESTANDAR'].unique()
paises
```

```python
#Figura total o parcial: nota 7.0
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(111)


for p in paises:
    dp=df[df['País__ESTANDAR']==p]
    dp=dp.sort_values(by='Años__ESTANDAR')
    ax.plot(dp['Años__ESTANDAR'],dp['value'],'.-',label=p)
    
ax.set_xlabel('Año')
ax.set_ylabel('PIB total anual(USD/persona)')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5));
fig.savefig('figura.jpg')
```
