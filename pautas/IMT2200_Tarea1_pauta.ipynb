### **f)** ¿En qué fecha se reportó el máximo número de nuevos casos confirmados a nivel nacional (máxima diferencia respecto al reporte anterior)?<br>

```python
dat=pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto1/Covid-19.csv')
dat.columns
```

```python
fechas=dat.columns[5:-1]
fechas
```

```python
maxnew=0
maxdate=0

for f in range(1,len(fechas)):
    new=dat[fechas[f]].sum()-dat[fechas[f-1]].sum()
    
    if new>maxnew:
        maxnew=new
        maxdate=fechas[f]
print(maxdate,maxnew)
```

### 3.5  Elija una comuna cualquiera de Chile (indique cuál) y genere un (único) gráfico con la evolución en el tiempo del los casos  confirmados incremental y de la movilida urbana, para todo el rango de fechas disponible. Para el gráfico de movilidad urbana, incluya barras de error para representar las cotas inferior y superior.



```python
comuna=13201.0 	#Puente alto

dat=pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto1/Covid-19.csv')

dat_covid=dat[dat['Codigo comuna']==comuna]
dat_covid
```

```python
from datetime import date

date_covid=dat_covid.columns[5:-1]
```

```python
date_covid=pd.to_datetime(pd.Series(date_covid))
```

```python
value_covid=dat_covid.iloc[0][5:-1]
```

```python
dat_mov=week[week['comuna']==comuna]
dat_mov
```

```python
fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(211)
ax2 = ax1.twinx()

ax1.plot(date_covid,value_covid,color='r',marker='.');
ax1.set_ylabel('Casos confirmados COVID',color='r',)

yerr_low=dat_mov['var_salidas']-dat_mov['var_salidas_cota_inferior']
yerr_up=dat_mov['var_salidas_cota_superior']-dat_mov['var_salidas']
ax2.plot(pd.to_datetime(dat_mov['fecha_inicio']),dat_mov['var_salidas'],'.',color='b')
ax2.errorbar(pd.to_datetime(dat_mov['fecha_inicio']),y=dat_mov['var_salidas'],
             yerr=[yerr_low,yerr_up],color='b',marker='.',ls='None')

ax2.set_ylabel('Movilidad',color='b')
ax1.set_xlabel('Fecha')
```
