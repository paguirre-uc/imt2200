¿Cómo se producen los datos faltantes? por ej, en encuestas, no toda la gente responde. Ademas, no todas las preguntas de las encuestas son obligatorias. 
En otros contextos, puede ser por error humano, falla en algún sistema de recolección de info. (por ej. un sensor), corrupción de datos, etc.
Los datos faltantes son relevantes en el análisis de datos, ya que puede generar sesgos. Los datos faltantes pueden ser totalmente al azar o puede haber un patrón detrás. 

2 principales soluciones: Eliminar filas con datos faltantes, o rellenarlas con algún valor, como por ejemplo el promedio en la columna (generalmente es mejor rellenar que eliminar). Esto se puede extender y además de rellenar las celdas con datos faltantes, señalamos las filas que tienen datos faltantes.
Para el ejemplo de las encuestas con preguntas opcionales: ¿tiene sentido eliminar las filas con datos faltantes? NO, generamos un sesgo a las personas que respondieron todo.

# Datos faltantes

En pandas, missing values generalmente se representan como NaN: not a number

```python
# ¿Son NaNs?
apps.isna()
# Exactamente la misma funcion:
# apps.isnull()
```

```python
print("N filas:", apps.shape[0])

# Función útil para ver los NO missing values por columna.
apps.info()
```

```python
# Como manejarlos? 
# Opcion 1: Borrar todas las filas con NaNs
apps.dropna()
# equivalente: apps.dropna('index')
```

```python
# Opcion 2: Borrar todas las columnas con NaNs
apps.dropna(axis=1)
# equivalente: apps.dropna('columns')
```

```python
# Opcion 3: Eliminar segun criterio especifico
# por ej, eliminar filas con mas de K NaNs
k = 4
apps.dropna(thresh=apps.shape[1]-k)
```

```python
# Tambien podemos borrar las columnas con mas de P% datos faltantes (para este ej p=15%)
p = 0.1
thresh = int((1-p)*apps.shape[0])
print('Se borraran las columnas con <', thresh, 'datos NO nulos')
#print(apps.info())
apps.dropna(axis=1, thresh=thresh)#.info()
```

```python
# Eliminar segun columnas especificas
apps.dropna(subset=['Genres'])
```

```python
# Opcion 4: Rellenar celdas con algun valor.
# Por ejemplo, intentemos rellenar Nans en 'Price' puede ser 0 o la media
# Para usar la media, tendremos que trabajar un poco la columna 'Price'

# Valores unicos
print(apps['Price'].unique(), '\n')
# lo siguiente se cae si lo intentan correr, ya que contiene strings del tipo '$x.xx' o NaNs
# apps['Price'].mean()

# Cantidad
print(apps['Price'].value_counts())
```

```python
# Opcion 1: Rellenar price con ceros
apps['Price'].fillna(0)
```

```python
# Opcion 2: Rellenar 'Price' con la media
def price_without_dollar(x):
    if '$' in x:
        return float(x[1:])
    return 0

price_mean = apps['Price'].map(price_without_dollar, na_action='ignore').dropna().mean()
print(f"Promedio de precios: ${price_mean:.3f}")
apps['Price'].fillna(price_mean)
```

```python
# Tambien podemos rellenar multiples columnas con un diccionario
values = {'Price': price_mean, 'Content Rating': 'Sin información'}
apps.fillna(value=values)
```

# Datos duplicados

```python
# Contar la cantidad de filas con datos duplicados
apps.duplicated().value_counts()
```

```python
# Ver cuales son las filas con datos duplicados
apps[apps.duplicated()]
```

```python
apps[apps['App']=='Toy Guitar with songs for kids']
```

```python
# Al igual que para buscar NaNs, podemos buscar duplicados en columnas especificas
apps.duplicated('App').value_counts()
# notar que esto igual maneja NaNs como iguales
# apps.duplicated(subset=['X', 'Y]).value_counts() # para varias columnas
```

¿Siempre es bueno borrar los duplicados? Depende del contexto. ¿Tiene sentido que hayan duplicados en nuestro dataset? Un ejemplo, algo burdo: Un dataset que contiene las ventas de una tienda en un día, que solo contiene el producto que fue vendido y su precio. Si eliminamos los duplicados, sería asumir que vendemos solo una unidad del producto durante el día. 

```python
# Para borrar duplicados
apps.drop_duplicates()
```

```python
# Al igual que para borrar NaNs, podemos especificar columnas para buscar duplicados
apps.drop_duplicates('App')
```

```python
apps[apps['App']=='CW Mobile']
```

```python
# Podemos elegir con cual duplicado quedarnos con el parametro 'keep' 
# {'first', 'last', False}, default ‘first’. False elimina los dos
apps.drop_duplicates(keep=False)
```
