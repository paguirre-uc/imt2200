### 2. Análisis de datos de ventas en locales de una cadena de retail.

Suponga que ud. tiene datasets correspondientes a los clientes de una tienda de retail, y todas las compras que ellos han realizado en distintos locales: `compras.csv`y `clientes.csv`.<br>

Además, cuenta con dos capas de datos vectoriales con la ubicación de los locales, y la zonificación de la ciudad.

Combine los datasets de manera adecuada para generar DataFrames que le permitan analizar las siguientes preguntas.

#### ¿Cuánto es el monto total comprado por cada cliente registrado?

```python
clientes=pd.read_csv('clientes.csv')
clientes.drop(columns='Unnamed: 0',inplace=True)
clientes
```

```python
compras=pd.read_csv('compras.csv')
compras.drop(columns='Unnamed: 0',inplace=True)

compras
```

```python
compras_clientes=clientes.merge(compras,on='Rut',how='inner')
compras_clientes.head()
```

```python
compras_clientes.groupby(by=['Rut'],as_index=False).Monto.sum()
```

```python
compras_clientes.groupby(by=['Rut','Local']).agg({'Edad':'min','Monto':['sum','mean']})
```

```python
def as_perc(value, total):
    return value/float(total)
```

```python
total_ventas=compras_clientes.Monto.sum()
compras_clientes[['Rut','Monto']].groupby(by=['Rut']).apply(as_perc,total=total_ventas)
```

```python
pv=pd.pivot_table(compras_clientes,index='Rut',aggfunc=np.std,values='Monto')
```

#### ¿Hay una relación entre el tipo de zona donde se ubica un local, y las ventas del local?

```python
import geopandas as gpd
```

```python
locales=gpd.read_file('locales.gpkg')
locales.crs
```

```python
zonas=gpd.read_file('zonas.gpkg').to_crs(4326)
zonas.head()
```

```python
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111)

zonas.plot(ax=ax,column='Tipo_zona',categorical=True,legend=True)
locales.plot(ax=ax,color='r')
```

```python
locales_zonas=gpd.sjoin(locales,zonas,how='left',predicate='intersects',rsuffix='zona')
locales_zonas
```

```python
compras_clientes['ID_local']=compras_clientes['Local'].str.replace('L','').astype('int')
compras_clientes.head()
```

```python
df=compras_clientes.merge(locales_zonas[['ID_local','Tipo_zona']],on='ID_local',how='left')
df
```

```python
pv=pd.pivot_table(df,index='Edad',columns='Tipo_zona',values='Monto',aggfunc='mean')
pv
```
